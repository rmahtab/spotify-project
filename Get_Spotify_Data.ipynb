```python
import pandas as pd

df = pd.read_csv('top-50-USA-cumulative.csv')
```

```python
df
```

```python
def format_date(date):
    nums = date.split("/")
    
    for i in range(len(nums)):
        if len(nums[i]) == 1:
            nums[i] = '0' + nums[i]
    
    return "/".join(nums)

df['Added'] = df['Added'].astype('str')
df['Removed'] = df['Removed'].astype('str')
df['Added'] = df['Added'].apply(format_date)
df['Removed'] = df['Removed'].apply(format_date)
df['Added'] =  pd.to_datetime(df['Added'], format='%m/%d/%y')
df['Removed'] =  pd.to_datetime(df['Removed'], format='%m/%d/%y')
```
